## 2D Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# 2D mean vector
mu = np.array([0, 0])
S = np.array([[1, 0], [0, 1]])
random_points = np.random.multivariate_normal(mu, S, 1000)

# Plot the points
plt.figure(figsize=(8, 8))
plt.scatter(random_points[:, 0], random_points[:, 1], alpha=0.6)
plt.title("2D Gaussian Distribution")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## 2D Laplacian

```python
mu = 0
beta = 1

x_laplace = np.random.laplace(mu, beta, 1000)
y_laplace = np.random.laplace(mu, beta, 1000)

random_points_laplace = np.column_stack((x_laplace, y_laplace))

# Plot the points
plt.figure(figsize=(8, 8))
plt.scatter(random_points_laplace[:, 0], random_points_laplace[:, 1], 
            alpha=0.6, color='purple', label='2D Laplace')
plt.title("2D Laplacian Distribution (Independent Components)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, alpha=0.3)
plt.show()
```
