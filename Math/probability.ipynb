# Probability theory in machine learning

## Expectation
**Expected value**
- The average, or mean value, that $f(x)$ takes on when $x$ is drawn from $P$
- For discrete values:
$$
\mathbb{E}_{x \sim P} \left[ f(x) \right] = \sum_x P(x)f(x)
$$
- For continous values:
$$
\mathbb{E}_{x \sim P} \left[ f(x) \right] = \int p(x)f(x) dx
$$

## Gaussian/Normal distribution
* $\mathbb{E}[x] = \mu$, where $\mu \in \mathbb{R}$
* $\sigma$ is standard deviation, where $\sigma \in (0,\infty)$
$$
\mathcal{N}(x,\mu,\sigma^2) = \sqrt{\frac{1}{2\pi \sigma^2}} \text{exp}\left({-\frac{1}{2 \sigma^2}(x-\mu)^2}\right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

x = np.linspace(-5, 5, 100)
mean = 0
std_dev = 1
y = gaussian(x, mean, std_dev)

area_under_curve = np.sum(y) * (x[1] - x[0])
print(f"Area under the curve: {area_under_curve}")

# Plot the Gaussian distribution
plt.plot(x, y, label='Gaussian distribution')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gaussian distribution')
plt.legend()
plt.show()
```

## Multivariate normal distribution
* Normal distribution generalized to $\mathbb{R}^n$
* $S$ is a positive symmetric matrix that gives the covariance of the distribution
* $\overline{\mu}$ gives the mean of the distribution (vector-valued mean)
$$
\mathcal{N}(\overline{x}, \overline{\mu}, S) = \sqrt{\frac{1}{(2\pi)^n \det(S)}} \text{exp} \left( -\frac{1}{2}(\overline{x}-\overline{\mu})^T S^{-1} (\overline{x}-\overline{\mu}) \right)
$$
