## Matrices

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1,2],[0,1]])
B = np.array([[2,0],[3,4]])

circle_points = np.linspace(0, 2*np.pi, 50)
x = np.cos(circle_points)
y = np.sin(circle_points)

transformed_points_a = A @ np.array([x, y])
transformed_points_b = B @ np.array([x, y])

# Plotting the points
plt.scatter(x, y, color='blue')
plt.scatter(transformed_points_a[0], transformed_points_a[1], color='red')
plt.scatter(transformed_points_b[0], transformed_points_b[1], color='green')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal')
plt.show()
```

## Norms

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a grid of points
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, y)

# Define the Norm functions
l1_norm = np.abs(X) + np.abs(Y)
l2_norm = np.sqrt(X**2 + Y**2)
linf_norm = np.maximum(np.abs(X), np.abs(Y))

# Plotting
plt.figure(figsize=(8, 8))

# Contour plots where the norm value is exactly 1.0
plt.contour(X, Y, l1_norm, levels=[1.0], colors='red', linewidths=2)   # L1
plt.contour(X, Y, l2_norm, levels=[1.0], colors='blue', linewidths=2)  # L2
plt.contour(X, Y, linf_norm, levels=[1.0], colors='green', linewidths=2) # L_inf

# Formatting to match your worksheet's intent
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(['L1 (Diamond)', 'L2 (Circle)', 'L-inf (Square)'], loc='upper right')
plt.title("Unit Circles for Different Norms ($||x||_p = 1$)")
plt.gca().set_aspect('equal')
plt.show()
```
